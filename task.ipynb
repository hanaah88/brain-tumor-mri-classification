```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from PIL import Image 
import os
```

```python
def load_images(folder , label , img_size=(64,64)):
    images=[]
    labels=[]
    for filename in os.listdir(folder):
        img_path = os.path.join(folder , filename)
        try:
            img = Image.open(img_path).convert("L")
            img = img.resize(img_size)
            img_array = np.array(img).flatten()
            images.append(img_array)
            labels.append(label)
            print(img)
        except:
            continue
    return images , labels
```

```python
notumor_imgs ,notumor_labels = load_images(r"D:\brain tumor\Training\notumor" , 0)
glioma_imgs ,glioma_labels = load_images(r"D:\brain tumor\Training\glioma" , 1)
meningioma_imgs ,meningioma_labels = load_images(r"D:\brain tumor\Training\meningioma" , 2)
pituitary_imgs ,pituitary_labels = load_images(r"D:\brain tumor\Training\pituitary" , 3)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA

X = notumor_imgs + glioma_imgs + pituitary_imgs + meningioma_imgs
y = notumor_labels + glioma_labels + pituitary_labels + meningioma_labels

X_train , X_test , y_train , y_test = train_test_split( X , y , train_size= 0.8 , random_state= 42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train , y_train)
y_pred2 = model_rf.predict(X_test)
print(f"training acc : {accuracy_score(y_train , model_rf.predict(X_train))}")
print(f"Testing acc : {accuracy_score(y_test , y_pred2)}")
print(confusion_matrix(y_test , y_pred2))

model_svm = SVC()
model_svm.fit(X_train , y_train)
y_pred3 = model_svm.predict(X_test)
print(f"training acc : {accuracy_score(y_train , model_svm.predict(X_train))}")
print(f"Testing acc : {accuracy_score(y_test , y_pred3)}")
print(confusion_matrix(y_test , y_pred3))
```

```python
def test_img(img_path , model):
    img = Image.open(img_path).convert("L")
    img = img.resize((64,64))
    img_array = np.array(img).flatten().reshape(1,-1)
    img_scaled = scaler.transform(img_array)
    prediction = model.predict(img_scaled)
    return prediction
```

```python
test_img(r"D:\brain tumor\Training\notumor\Te-no_2.jpg" , model_svm)
```
